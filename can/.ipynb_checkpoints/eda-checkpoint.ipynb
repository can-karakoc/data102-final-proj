```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import re
```

```python
electricity = pd.read_csv('../cleaned_datasets/electricity_cleaned.csv')
electricity.sort_values('Average retail price of electricity', ascending=False).head()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
sns.histplot(electricity['Average retail price of electricity'], stat='density', bins=30, ax=ax, 
             color='dodgerblue', edgecolor='steelblue')
plt.title('Average Annual Retail Price of Electricity by State')
plt.xlabel('Price (cents for kilowatthour)')
plt.show()
```
