```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('datasets/state_political_party1.csv')
data.head()
```

```python
political_affiliation_counts = data['R or D'].value_counts()

plt.figure(figsize=(8, 6))
political_affiliation_counts.plot(kind='bar', color=['blue', 'red'], alpha=0.7)
plt.title('State Political Leanings (2020 Election)', fontsize=14)
plt.xlabel('Political Affiliation', fontsize=12)
plt.ylabel('Number of States', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

political_affiliation_counts
```
