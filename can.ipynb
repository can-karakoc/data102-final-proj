```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import re
```

## Data Cleaning for Net Summer Capacity Using Primarily Renewable Energy Sources by State Data

Steps to clean the data and reformat the table:
- The dataset includes 'NM' not meaningful due to large relative standard error. The 'NM' entries were converted to NaN.
- The data were in string format, thus converted the data to float.

```python
capacity = pd.read_csv('datasets/net_summer_capacity.csv', header=[1, 2], skiprows=[3])

# Assign "NM" -> NaN and change the data to float
numeric_cols = capacity.columns[1:]
capacity = capacity.replace("NM", np.nan) 
capacity[numeric_cols] = capacity[numeric_cols].replace(',', '', regex=True).astype(float)
capacity.rename(columns=lambda x: x.replace('\n', ' '), inplace=True)

capacity = capacity.loc[:, capacity.columns.get_level_values(0)]
capacity.columns = [' '.join(col).strip() for col in capacity.columns.values]

capacity = capacity.iloc[:-1]
```

```python
def rename_columns(cols):
    new_cols = []
    new_cols.append(cols[0])
    change_cols = cols[1:]
    for round in range(1, len(change_cols)+1):
        # in odd turns, record name
        if round % 2 != 0:
            rename = re.match(r"^(.*?)(?= September)", cols[round])
            new_cols.append(cols[round])
        # in even rounds, rename
        elif round % 2 == 0:
            new_name = re.sub(r"^(.*?)(?= September)", rename.group(0), cols[round])
            new_cols.append(new_name)
    return new_cols

new_cols = rename_columns(capacity.columns)
capacity.columns = new_cols
capacity.head()
```

The data contains the net summer capacity (Megawatts) using primarily renewable energy sources and by state in September 2024 and 2023. Since our analysis is focused on 2023, we will only consider 2023 data. As a result, we only considered columns measured in September 2023.

```python
capacity2023 = capacity[capacity.columns[~capacity.columns.str.contains('2024')]]
capacity2023.head()
```

To be able to make more clear visualizations, we dropped "September 2023" from the columns.

```python
capacity2023.columns = capacity2023.columns.str.replace('September 2023', '').str.strip()
capacity2023 = capacity2023.iloc[:, :7]
capacity2023
```

## EDA for Net Summer Capacity from Primarily Renewable Energy Sources by State in September 2023

```python
divisions = ["East North Central", "West North Central", 
                         "East South Central", "West South Central", 
                         "Pacific Contiguous", "Pacific Noncontiguous", "South Atlantic", "Middle Atlantic", "Mountain", "New England"]

capacity2023_states = capacity2023[~capacity2023['Census Division and State'].isin(divisions)]
capacity2023_states = capacity2023_states.rename(columns={'Census Division and State': "State"})

capacity2023_states.iloc[:-1].sort_values(by=['State']).set_index('State').plot(kind='bar', stacked=True, figsize=(15, 6))

plt.xticks(rotation=45, ha='right', fontsize=11)
plt.title('Renewable Energy Capacity by State', fontsize=16)
plt.ylabel('Capacity (MW)', fontsize=14)
plt.xlabel('State', fontsize=14)
plt.tight_layout()
```

```python
capacity2023_divisions = capacity2023[capacity2023['Census Division and State'].isin(divisions)]
capacity2023_divisions = capacity2023_divisions.rename(columns={'Census Division and State': "Division"})

capacity2023_divisions.iloc[:-1].sort_values('Division').set_index('Division').plot(kind='bar', stacked=True, figsize=(10, 6))

plt.xticks(rotation=45, ha='right', fontsize=11)
plt.title('Renewable Energy Capacity by Division', fontsize=16)
plt.ylabel('Capacity (MW)', fontsize=14)
plt.xlabel('Division', fontsize=14)
plt.tight_layout()
```

```python
capacity2023_melted = pd.melt(capacity2023_states.iloc[:-1], id_vars=["State"], 
                    var_name="Energy Type", value_name="Capacity")
capacity2023_melted = capacity2023_melted.sort_values('State')
capacity2023_melted = capacity2023_melted[capacity2023_melted['Capacity'] > 0].dropna(subset=['Capacity'])
capacity2023_melted = capacity2023_melted.sort_values(['State', 'Capacity'])
```
